```python
import muspy
import os

def calc_metric(fname_iterationNo, path):
    fname = path + "/" + fname_iterationNo[0]
    print(fname)
    iterationNo = fname_iterationNo[1]
    music = muspy.read(fname)
    metrics = {}
    metrics["file_name"] = fname_iterationNo[0]
    metrics["iterationNo"] = iterationNo
    metrics["pitrch_range"] = muspy.pitch_range(music)
    metrics["n_pitches_used"] = muspy.n_pitches_used(music)
    metrics["n_pitch_classes_used"] = muspy.n_pitch_classes_used(music)
    metrics["polyphony"] = muspy.polyphony(music)
    metrics["polyphony_rate"] = muspy.polyphony_rate(music, threshold=2)
    metrics["scale_consistency"] = muspy.scale_consistency(music)
    metrics["pitch_entropy"] = muspy.pitch_entropy(music)
    metrics["pitch_class_entropy"] = muspy.pitch_class_entropy(music)

    metrics["empty_beat_rate"] = muspy.empty_beat_rate(music)
    metrics["drum_pattern_consistency"] = muspy.drum_pattern_consistency(music)
    metrics["groove_consistency"] = muspy.groove_consistency(music, measure_resolution=4)
    metrics["empty_measure_rate"] = muspy.empty_measure_rate(music, measure_resolution=4)
    return metrics
```

```python
from pathlib import Path

def get_metrics_of_musics(df, directory, iteration_no):

    # Create a Path object for the directory
    directory = Path(directory)

    # Use glob to find all files ending with ".mid"
    midi_files = directory.glob("*.mid")
    
    for file in midi_files:
        df = df.append(calc_metric((file.name,iteration_no), str(directory)), ignore_index=True)
    
    return df
    

# fname = "output_3000take0_beat_80.mid"
# dick = calc_metric(fname)
# dick.keys()
```
